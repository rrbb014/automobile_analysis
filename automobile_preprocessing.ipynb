### Remove what target variable(normalized-losses) is "?"

```python
for idx in range(DATA.shape[0]):
    if DATA.loc[idx]['normalized-losses'] == '?':
        print("idx : %d" % idx)
        print(DATA.loc[idx]['normalized-losses'])
        DATA = DATA.drop(idx)  
```
